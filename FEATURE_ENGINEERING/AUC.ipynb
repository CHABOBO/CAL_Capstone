```python
# Smaller file, but missing historical fields...
import pandas as pd
import numpy as np

df = pd.read_csv("dfaa.csv", index_col=0, parse_dates=True)
df.head()
```

```python
# the entire dataset
import pandas as pd
import numpy as np

df = pd.read_csv("df_2.csv", index_col=0, parse_dates=True)
df.head()
```

```python
def check_readmission(input_str):
    input_str = str(input_str)
    if input_str.lower() == "true":
        return 1
    return 0

#Argh, how do you reference a field name starting with a number?
df["2m_readmission"] = 0
df.loc[:, '2m_readmission'] = df.2m_read.apply(check_readmission)
```

```python
#Sample
from sklearn import metrics

data = df[np.isfinite(df['avgbmi'])]

read=data["1m_read"]
feature= data["avgbmi"]

#print(read)
#print(feature)

'''calculating ROC AUC'''
fpr,tpr,thresholds = metrics.roc_curve(read,feature)
aucvalue=metrics.auc(fpr,tpr)
print('AUC value:')
print(aucvalue)
print(len(data.index))

```

```python
AUCs = pd.DataFrame(columns=["feature", "AUC", "Count"])

for field in df:    
    try:
        data = df[np.isfinite(df[field])]

        read = data["1m_read"]
        feature = data[field]

        '''calculating ROC AUC'''
        fpr,tpr,thresholds = metrics.roc_curve(read,feature)
        aucvalue=metrics.auc(fpr,tpr)
        
        rowcount = len(data.index)
        
        #print(field)
        #print(aucvalue)
        AUCs = AUCs.append({"feature": field, "AUC": aucvalue, "Count": rowcount}, ignore_index = True)
        
    except:
        pass

with pd.option_context('display.max_rows', None, 'display.max_columns', 3):    
    print(AUCs.sort_values(by=["AUC"], ascending=False))
```
