## This calculates the AUC for each field when predicting 1m_read

```python
# the entire dataset
import pandas as pd
import numpy as np

dfall = pd.read_csv("df_v4.csv", index_col=0, parse_dates=True)
dfall.head()
```

```python
dfall.shape
```

```python
#filter the dataset to the subset of interest...
df = dfall[dfall.maxdied_ever==False]
df = df[df.los_10day==True]

df.shape
```

```python
#Sample
from sklearn import metrics

data = df[np.isfinite(df['avgbmi'])]

read=data["2m_read"]
feature= data["avgbmi"]

#print(read)
#print(feature)

'''calculating ROC AUC'''
fpr,tpr,thresholds = metrics.roc_curve(read,feature)
aucvalue=metrics.auc(fpr,tpr)
print('AUC value:')
print(aucvalue)
print(len(data.index))

```

```python
AUCs = pd.DataFrame(columns=["feature", "AUC", "Count"])

for field in df:    
    try:
        data = df[np.isfinite(df[field])]

        read = data["1m_read"]
        feature = data[field]

        '''calculating ROC AUC'''
        fpr,tpr,thresholds = metrics.roc_curve(read,feature)
        aucvalue=metrics.auc(fpr,tpr)
        
        rowcount = len(data.index)
        
        #print(field)
        #print(aucvalue)
        AUCs = AUCs.append({"feature": field, "AUC": aucvalue, "Count": rowcount}, ignore_index = True)
        
    except:
        pass

with pd.option_context('display.max_rows', None, 'display.max_columns', 3):    
    print(AUCs.sort_values(by=["AUC"], ascending=False))
```
